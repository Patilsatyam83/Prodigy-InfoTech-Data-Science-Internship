Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV, skipping metadata rows (first 4 rows)
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_23043.csv', skiprows=4)

# Select relevant columns: Country Name and 2024 population
df = df[['Country Name', '2024']]

# Drop rows with missing population
df = df.dropna(subset=['2024'])

# Convert population to numeric
df['2024'] = pd.to_numeric(df['2024'], errors='coerce')
```

```python
top10 = df.sort_values('2024', ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10['2024'])
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Countries by Population in 2024')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['2024'], bins=20, edgecolor='black')
plt.title('Distribution of Country Populations in 2024')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```

## Solution Explanation

For this task, I visualized the distribution of the population variable from the World Bank dataset for the year 2024.

- **Bar Chart:** Shows the top 10 countries by population in 2024. This highlights which countries have the largest populations.
- **Histogram:** Shows how the populations of all countries are distributed. Most countries have relatively small populations, while a few have very large populations.

These visualizations help us quickly understand global population patterns and identify outliers.

## Insights

- The bar chart reveals that countries like China and India have the largest populations.
- The histogram shows that the majority of countries have populations below 100 million, with only a few exceeding this number.
